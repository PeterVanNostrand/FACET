```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
best_dists = pd.read_csv("dists_stack1.csv")
```

```python
xs = []
ys = []
for idx, row in best_dists.iterrows():
    times = []
    dists = []
    sample_bests = row.dropna()[1:]
    for item in sample_bests:
        t, d = item[1:-1].replace(" ", "").split(",")
        times.append(int(t))
        dists.append(float(d))
    xs.append(np.array(times))
    ys.append(np.array(dists))
```

```python
for x, y in zip(xs, ys):
    plt.plot(x, (y - np.min(y)) / (np.max(y) - np.min(y)))
plt.xlabel("Nodes Bounded")
plt.ylabel("Solution Distance (Normalize)")
plt.title("Solution Distance During Explanation")
plt.savefig("stack_dists.png")
```
