```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
k = 5
x1 = np.arange(k+1, 100)
x2 = np.arange(k+1, 100, dtype=int)
y1 = x1**k # n^k
y2 = np.zeros(x1.shape[0])
for i in range(x2.shape[0]):
    y2[i] = math.factorial(x2[i]) / (math.factorial(k) * math.factorial(x2[i] - k))
```

```python
plt.plot(x1, y1)
```

```python
plt.plot(x2, y2)
```

```python
import math
def nchoosek(n, k):
    return math.factorial(n) / (math.factorial(k) * math.factorial(n-k))

def n_regions(T, D):
    n_majority = math.ceil(T / 2)
    n_leaves = T * 2**D
    return nchoosek(n_leaves / 2, n_majority)
```

```python
n_trees = [20, 65, 150]
for t in range(10, 100):
    print(t, n_regions(t, 5))
```
