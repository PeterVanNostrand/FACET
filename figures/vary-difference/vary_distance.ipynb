```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cardio = pd.read_csv("../../results/vary-difference/cardio.csv")
musk = pd.read_csv("../../results/vary-difference/musk.csv")
thyroid = pd.read_csv("../../results/vary-difference/thyroid.csv")
wbc = pd.read_csv("../../results/vary-difference/wbc.csv")
```

```python
cmap = plt.get_cmap("Set1")
colors = cmap.colors
```

```python
# coverage
plt.figure()
plt.plot(cardio.difference, cardio.coverage_ratio, label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.difference, musk.coverage_ratio, label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.difference, thyroid.coverage_ratio, label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.difference, wbc.coverage_ratio, label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Difference Ratio")
plt.ylabel("Coverage Ratio")
plt.title("Coverage vs Difference")
plt.legend(loc="best")
plt.savefig("coverage_difference.png")
```

```python
# distance
plt.figure()
plt.plot(cardio.difference, cardio.mean_distance, label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.difference, musk.mean_distance, label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.difference, thyroid.mean_distance, label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.difference, wbc.mean_distance, label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Difference Ratio")
plt.ylabel("Mean Distance")
plt.title("Mean Distance vs Difference")
plt.legend(loc="best")
plt.savefig("distance_difference.png")
```

```python
# distance ratio
plt.figure()
plt.plot(cardio.difference, cardio.mean_distance / np.sqrt(21), label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.difference, musk.mean_distance / np.sqrt(166), label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.difference, thyroid.mean_distance / np.sqrt(6), label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.difference, wbc.mean_distance / np.sqrt(30), label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Difference Ratio")
plt.ylabel("Mean Distance Ratio")
plt.title("Mean Distance Ratio vs Difference")
plt.legend(loc="best")
plt.savefig("distance_ratio_difference.png")
```

```python
# f1 score
plt.figure()
plt.plot(cardio.difference, cardio.f1, label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.difference, musk.f1, label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.difference, thyroid.f1, label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.difference, wbc.f1, label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Difference Ratio")
plt.ylabel("F1 Score")
plt.title("F1 Score vs Difference")
plt.legend(loc="best")
plt.savefig("f1_difference.png")
```
