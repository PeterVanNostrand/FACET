```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cardio = pd.read_csv("../../results/vary-k/cardio.csv")
musk = pd.read_csv("../../results/vary-k/musk.csv")
thyroid = pd.read_csv("../../results/vary-k/thyroid.csv")
wbc = pd.read_csv("../../results/vary-k/wbc.csv")
```

```python
cmap = plt.get_cmap("Set1")
colors = cmap.colors
```

```python
# coverage
plt.figure()
plt.plot(cardio.k, cardio.coverage_ratio, label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.k, musk.coverage_ratio, label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.k, thyroid.coverage_ratio, label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.k, wbc.coverage_ratio, label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Num Candiates (k)")
plt.ylabel("Coverage Ratio")
plt.title("Coverage vs Num Candidates")
plt.legend(loc="best")
plt.savefig("coverage_k.png")
```

```python
# distance
plt.figure()
plt.plot(cardio.k, cardio.mean_distance, label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.k, musk.mean_distance, label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.k, thyroid.mean_distance, label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.k, wbc.mean_distance, label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Num Candiates (k)")
plt.ylabel("Mean Distance")
plt.title("Mean Distance vs Num Candidates")
plt.legend(loc="best")
plt.savefig("distance_k.png")
```

```python
# distance
plt.figure()
plt.plot(cardio.k, cardio.mean_distance / np.sqrt(21), label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.k, musk.mean_distance / np.sqrt(166), label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.k, thyroid.mean_distance / np.sqrt(6), label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.k, wbc.mean_distance / np.sqrt(30), label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Num Candiates (k)")
plt.ylabel("Mean Distance Ratio")
plt.title("Mean Distance Ratio vs Num Candidates")
plt.legend(loc="best")
plt.savefig("distance_ratio_k.png")
```

```python
# f1 score
plt.figure()
plt.plot(cardio.k, cardio.f1, label="Cardio", color=colors[0], lw=2.0)
plt.plot(musk.k, musk.f1, label="Musk", color=colors[1], lw=2.0)
plt.plot(thyroid.k, thyroid.f1, label="Thyroid", color=colors[3], lw=2.0)
plt.plot(wbc.k, wbc.f1, label="WBC", color=colors[4], lw=2.0)
plt.xlabel("Num Candiates (k)")
plt.ylabel("F1 Score")
plt.title("F1 Score vs Num Candiates")
plt.legend(loc="best")
plt.ylim([0,1])
plt.savefig("f1_k.png")
```
